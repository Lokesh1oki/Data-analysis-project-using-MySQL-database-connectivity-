```python
merge=pd.merge(models,performance,on='model_id',how='inner')          #joins two tables
```

```python
merge=pd.merge(merge,developers,on='model_id',how='inner')            #joins combined two tables with third
```

```python
merge                                                                 #prints merged tables
```

```python
import matplotlib.pyplot as plt                         #import matplotlib as plt
average=merge['score'].mean()                           #finds average of score
plt.figure(figsize=(10,7))                              #sets figure size for plot
plt.bar(merge['model_name'],average)                    #plots bar chart
plt.title('average permormance score by models')        #gives title
plt.xlabel('model name')                                #gives xlabel
plt.ylabel('aveage score')                              #gives ylabel 
plt.xticks(rotation='90')                               #xaxis label rotation
plt.show()                                              #displays chart
```

```python
merge['model_type'].value_counts().plot(kind='pie',autopct='%1.1f%%') #plots pie chart for count of model type
plt.title('model type division')                                      #gives title
```

```python
plt.hist(merge['score'],bins=10)            #plots the hhistogram chart
plt.title('score frequency')                #gives title
plt.xlabel('score')                         #gives xlabel
plt.ylabel('frequency')                     #gives ylabel
plt.show()                                  #used to display the chart
```

```python
date=merge['evaluation_date']                                #variable to store the date
s=merge['score']                                             #variable to store score
plt.plot(date,s)                                             #used to plot kine chart
plt.xticks(rotation=90)                                      #rotates the xlabel to 90 degree
plt.xlabel('evalution date')                                 #gives xlabel
plt.ylabel('score')                                          #gives ylabel
plt.title('score of ai model over datetime')                 #gives title
plt.show()                                                   #displays the chart
```

```python
plt.figure(figsize=(14,7))                             #sets figure size for plot
plt.scatter(merge['developer_name'],merge['score'])    #plots the scatter chart
plt.xlabel('developer names')                          #gives xlabel
plt.ylabel('their model scores')                       #gives ylabel
plt.title('developers name and thier project scores')  #gives title
plt.xticks(rotation=75)                                #rotates xlabel to 75 degree
plt.show()                                             #displays the chart
```
